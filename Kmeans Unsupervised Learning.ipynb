```python
# Unsupervised no sampling
```

```python
from sklearn.cluster import KMeans
```

```python
K=KMeans(n_clusters=4)
```

```python
K.fit(a)
```

```python
# IF 4 Clusters have been created so each record have gone into one of the cluster
```

```python
K.labels_
```

```python
len(K.labels_)
# for each record u have 1 label coz its on one mapping
```
